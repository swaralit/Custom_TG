# Drug reference table

## Statins

```python
%%read_sql

create or replace temporary table st_ndc_ref1 as

select
    'ndc' as cat1,
    'statin' as cat2,
    null as cat3,
    ndcnum as value,
    THRCLDS as description,
    gennme as generic_name,
    current_date() as active_date,
    'swar' as active_reason,
    null as deactive_date,
    null as deactive_reason,
    'RWD_DB.RWD_REFERENCE_LIBRARY.AMN_REDBOOK_LATEST' as source
from 
   RWD_DB.RWD_REFERENCE_LIBRARY.AMN_REDBOOK_LATEST
    where
        thrdtds ilike '%statin%'
        and thrclds ilike '%antihyperlipidemic%'
        and not (thrdtds ilike '%atorvastatin%'
        or thrdtds ilike '%fluvastatin%'
        or thrdtds ilike '%lovastatin%'
        or thrdtds ilike '%pravastatin%'
        or thrdtds ilike '%rosuvastatin%'
        or thrdtds ilike '%simvavastatin%'
        or thrdtds ilike '%pitavastatin%')
```

```python
snow.select("select * from st_ndc_ref1")
```

```python
%%read_sql

create or replace temporary table st_ndc_ref2 as

select
    'ndc' as cat1,
    'statin' as cat2,
    case 
        when thrdtds ilike '%Atorvastatin%' then 'Atorvastatin'
        when thrdtds ilike '%Fluvastatin%' then 'Fluvastatin'
        when thrdtds ilike '%lovastatin%' then 'lovastatin'
        when thrdtds ilike '%pravastatin%' then 'pravastatin'
        when thrdtds ilike '%rosuvastatin%' then 'rosuvastatin'
        when thrdtds ilike '%simvastatin%' then 'simvastatin'
        when thrdtds ilike '%pitavastatin%' then 'pitavastatin'
       end as cat3,
    ndcnum as value,
    THRCLDS as description,
    gennme as generic_name,
    current_date() as active_date,
    'swar' as active_reason,
    null as deactive_date,
    null as deactive_reason,
    'RWD_DB.RWD_REFERENCE_LIBRARY.AMN_REDBOOK_LATEST' as source
from 
   RWD_DB.RWD_REFERENCE_LIBRARY.AMN_REDBOOK_LATEST
    where
        thrclds ilike '%antihyperlipidemic%'
        and (thrdtds ilike '%atorvastatin%'
        or thrdtds ilike '%fluvastatin%'
        or thrdtds ilike '%lovastatin%'
        or thrdtds ilike '%pravastatin%'
        or thrdtds ilike '%rosuvastatin%'
        or thrdtds ilike '%simvastatin%'
        or thrdtds ilike '%pitavastatin%') 
```

```python
snow.select("select * from st_ndc_ref2")
```

## Selecting records corresponding to NDC codes
Variables in pharmacy table: claim_id, PRODUCT_OR_SERVICE_ID, PRODUCT_OR_SERVICE_ID_QUALIFIER(use 03 code for NDC codes)

```python
%%read_sql

create or replace table st_ndc_all as

select
    patient_id,
    claim_number,
    diagnosis_code,
    date_of_service,
    PRODUCT_OR_SERVICE_ID_QUALIFIER,
    PRODUCT_OR_SERVICE_ID 

from
    RWD_DB.RWD.RAVEN_PHARMACY
where
    PRODUCT_OR_SERVICE_ID_QUALIFIER in ('03')
   --and 
    --PRODUCT_OR_SERVICE_ID in (select distinct value from st_ndc_ref)
    --and PRODUCT_OR_SERVICE_ID not like ('xxxxx-xxxx-xx')
```

```python
%%read_sql

begin;
update st_tg_hdl_agg1 a
set a.patient_id = b.patient_id
from st_ndc_all b
--where a.claim_number = b.claim_number
```

```python
%%read_sql

select * from st_tg_hdl_agg1 limit 20
```

```python
%%read_sql

create or replace table st_tg_ndc_agg1 as 
select a.patient_id,
        a.obs_quan,
     a.tg_level,
     a.unit,
     b.claim_number,
     b.product_or_service_id,
     b.date_of_service
 from
    st_tg_hdl_agg1 a
        join st_ndc_all b
            on a.patient_id = b.patient_id
            
            where left (a.patient_id,5) != ('XXX -')
--group by 1
```
